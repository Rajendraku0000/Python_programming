```python
import numpy as np
import pandas as pd
```

```python
import requests
```

```python
from bs4 import  BeautifulSoup
```

```python
url="https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
data=requests.get(url)
new_data=data.content
```

```python
soup=BeautifulSoup(new_data,"html.parser")
```

```python
company_name=soup.select("._4rR01T")
company=[]
for i in company_name:
#     print(i.contents[0])
    company.append(i.contents[0])
```

```python
price_list=soup.select("._30jeq3")
price=[]
for i in price_list:
#     print(i.contents[0])
    price.append(i.contents[0])
```

```python
review_list=soup.select("._3LWZlK")
review=[]
for i in review_list:
#     print(i.contents[0])
    review.append(i.contents[0])
```
